Question 1

You need to create an application that calculates the interest earned for each year for three
consecutive years. Interest is earned at 4% each year, hence the savings will increase by said 4%
for each year. Your goal is to develop the program to reflect this, with an output that looks
exactly as seen below – please note the “green text” are inputted information that the end-user
enters

```python
print("Welcome to your new premium savings account")

# Get name, PIN, current year, and initial deposit
name = input("Please enter your name: ")
pin = int(input("Please enter the PIN number associated with your account: "))
current_year = int(input("Please enter the current year: "))
balance = float(input("Please enter a savings amount: "))

# Constant interest rate
INTEREST_RATE = 0.04  # 4% annual interest

# Function to print summary
def print_summary(year, starting_balance, earned_interest, current_bal, i):
    print("---------------------------")
    print(f"Year: {year}")
    # If starting year:
    if i == 0:
        print(f"Original Savings: {starting_balance}")
    else:
        print(f"Previous Savings: {starting_balance:.2f}")
    print(f"Earned Interest: {earned_interest:.2f}")
    print(f"Current Savings: {current_bal:.2f}")
    print("---------------------------")

# Function to calculate the earned interest and current savings
def calculate_interest(balance, INTEREST_RATE):
    earned_interest = balance * INTEREST_RATE
    current_bal = balance + earned_interest
    return earned_interest, current_bal

# Loop through 3 consecutive years
for i in range(0, 3):
    earned_interest, balance = calculate_interest(balance, INTEREST_RATE)
    print_summary(current_year + i, balance - earned_interest, earned_interest, balance, i)
```

Question 2

Create a program for a hotel booking system. The first line of the program should have a title statement (e.g., “X’s hotel booking system” or something similar). The standard room for a night cost $79.99 (specifically, this would be a constant). The program, however, should accept the user’s input as to the number of nights the user will be staying. Hence, the program would calculate the guest’s hotel cost, which is based on the cost for a night by the number of nights they stay at the hotel. Additionally, users may have one of four membership types: (1) Basic, (2) Advanced, (3) Premium, and (4) VIP. These memberships allow for discounts of 10%, 15%, 25% and 40% respectively. However, not all guests would be members, and thus, would pay the full price.

```python
print("Welcome to Amaranth Hotel.")

# Take guests name and number of nights staying
name = input("Guest Name: ")
nights = int(input("Number of nights staying: "))

# Print a list of membership types
print("Membership Types:")
print("(1) Non-Member - NM")
print("(2) Basic - B")
print("(3) Advanced - A")
print("(4) Premium - P")
print("(5) VIP - V")

# Take membership input & standardize
membership = input("Please enter membership type: ").strip().upper()

# Price of stay ($79.99 * num of nights)
NIGHTLY_RATE = 79.99
initial_price = NIGHTLY_RATE * nights

# Discounted price by member type
if membership == 'NM':
    price = initial_price 
elif membership == 'B':
    price = initial_price * .90 
elif membership == 'A':
    price = initial_price * .85
elif membership == 'P':
    price = initial_price * .75
elif membership == 'V':
    price = initial_price * .60
else:
    print("Invalid input for membership.")

# Print Summary
print(f"Name: {name}")
print(f"Number of Nights: {nights}")
print(f"Membership Type: {membership}")
print(f"Price: {initial_price:.2f}")
print(f"Discounted Price: {price:.2f}")
```

Question 3

Copy the code from the previous hotel – in here, you will alter it to add the following:

Instead of a standard price of $79.99 for a room, there are:

Single Room (1 Double Bed): $59.99
Double Room (2 Double Beds): $79.99
Single Queen Room (1 Queen Bed): $89.99
Double Queen Room (2 Queen Beds): $99.99
Single King Room (1 King Bed): $110.99
Double King Room (2 King Beds): $139.99
Additionally, for the Double King Room, the user can request a sofa-couch, which will be an additional $10.00. The memberships are the same.

```python
print("Welcome to Amaranth Hotel.")

# Take guests name and number of nights staying
name = input("Guest Name: ")
nights = int(input("Number of nights staying: "))
# Display room types and prices
print("Here are the available room types:\n"
      "    1. Single Room (1 Double Bed): $59.99\n"
      "    2. Double Room (2 Double Beds): $79.99\n"
      "    3. Single Queen Room (1 Queen Bed): $89.99\n"
      "    4. Double Queen Room (2 Queen Beds): $99.99\n"
      "    5. Single King Room (1 King Bed): $110.99\n"
      "    6. Double King Room (2 King Beds): $139.99")

# Room selection
room_type = int(input("Enter your selected room type, using the number on the list: "))

# Room types & rates dictionary
room_rates = {
    1: 59.99,
    2: 79.99,
    3: 89.99,
    4: 99.99,
    5: 110.99,
    6: 139.99
}

# Validate room type
if room_type not in room_rates:
    print("Invalid room type selected. Defaulting to Double Room ($79.99).")
    nightly_rate = 79.99
else:
    # Dictionary makes this a lot easier
    nightly_rate = room_rates[room_type]

# Ask if they want a sofa
sofa = input("Would you like a sofa in the room (extra $10.00)? Enter Y or N: ")
sofa = sofa.lower().strip() # Standardize

# Membership types
print("Membership Types:")
print("(1) Non-Member - NM")
print("(2) Basic - B")
print("(3) Advanced - A")
print("(4) Premium - P")
print("(5) VIP - V")

# Take membership input & standardize
membership = input("Please enter membership type: ").strip().upper()

# Calculate base price
initial_price = nightly_rate * nights

# Add sofa fee
if sofa == "y":
    initial_price += 10 * nights

# Discounted price by member type
if membership == 'NM':
    price = initial_price 
elif membership == 'B':
    price = initial_price * .90 
elif membership == 'A':
    price = initial_price * .85
elif membership == 'P':
    price = initial_price * .75
elif membership == 'V':
    price = initial_price * .60
else:
    print("Invalid input for membership.")

# Print Summary
print(f"Name: {name}")
print(f"Number of Nights: {nights}")
print(f"Membership Type: {membership}")
print(f"Price: {initial_price:.2f}")
print(f"Discounted Price: {price:.2f}")
```

Question 4

Copy and paste (i.e., make another instance of this program) again which alters it even further – this time, rather than simply making the calculations, you must store all the necessary information (either entered into the program from the end-user or calculated) into lists – and you should be able to enter as many people in the program as need be. You will, of course, need a prompt as to whether the end-user wants to continue adding new check-ins or not….ultimately, when all is done, a report of each check-in (such as guest name, their choice of suite, prices, etc.) should be produced.

```python
print("Initializing Guest List for Amaranth Hotel...")

# Create master list of guests
guest_list = []

print("Beginning Guest Check in...")

# Loop through check in procedure until user terminates the loop
cont = True
while cont:
    print("\nWelcome to Amaranth Hotel.")
    
    # Take guest name and number of nights staying
    name = input("Guest Name: ")
    nights = int(input("Number of nights staying: "))
    
    # Display room types and prices
    print("Here are the available room types:\n"
          "    1. Single Room (1 Double Bed): $59.99\n"
          "    2. Double Room (2 Double Beds): $79.99\n"
          "    3. Single Queen Room (1 Queen Bed): $89.99\n"
          "    4. Double Queen Room (2 Queen Beds): $99.99\n"
          "    5. Single King Room (1 King Bed): $110.99\n"
          "    6. Double King Room (2 King Beds): $139.99")

    # Room selection
    room_type = int(input("Enter your selected room type, using the number on the list: "))

    # Room types & rates dictionary
    room_rates = {
        1: 59.99,
        2: 79.99,
        3: 89.99,
        4: 99.99,
        5: 110.99,
        6: 139.99
    }

    # Validate room type
    if room_type not in room_rates:
        print("Invalid room type selected. Defaulting to Double Room ($79.99).")
        nightly_rate = 79.99
    else:
        nightly_rate = room_rates[room_type]

    # Ask if they want a sofa
    sofa = input("Would you like a sofa in the room (extra $10.00)? Enter Y or N: ").lower().strip()

    # Membership types
    print("Membership Types:")
    print("(1) Non-Member - NM")
    print("(2) Basic - B")
    print("(3) Advanced - A")
    print("(4) Premium - P")
    print("(5) VIP - V")

    # Take membership input & standardize
    membership = input("Please enter membership type: ").strip().upper()

    # Calculate base price
    initial_price = nightly_rate * nights

    # Add sofa fee
    if sofa == "y":
        initial_price += 10 * nights

    # Discounted price by member type
    if membership == 'NM':
        price = initial_price
    elif membership == 'B':
        price = initial_price * 0.90
    elif membership == 'A':
        price = initial_price * 0.85
    elif membership == 'P':
        price = initial_price * 0.75
    elif membership == 'V':
        price = initial_price * 0.60
    else:
        print("Invalid input for membership. No discount applied.")
        price = initial_price

    # Print Summary for this guest
    print("\n--- Guest Check-In Summary ---")
    print(f"Name: {name}")
    print(f"Number of Nights: {nights}")
    print(f"Membership Type: {membership}")
    print(f"Base Price: ${initial_price:.2f}")
    print(f"Final Price: ${price:.2f}")

    # Save guest info to master list
    guest_list.append({
        "name": name,
        "nights": nights,
        "membership": membership,
        "room_type": room_type,
        "sofa": sofa,
        "base_price": initial_price,
        "final_price": price
    })
    
    # Check if user wants to loop again
    while True:
        check = input("Would you like to check in another guest (Y/N)?: ").lower().strip()
        if check == 'y':
            cont = True   # keep looping
            break
        elif check == 'n':
            cont = False  # stop looping
            break
        else:
            print("Invalid answer. Please enter Y or N.")

# Print Guest List Summary
print("\n=== Final Guest List Summary ===")
for guest in guest_list:
    print(f"{guest['name']} | {guest['nights']} nights | "
          f"Membership: {guest['membership']} | "
          f"Room Type: {guest['room_type']} | "
          f"Sofa: {guest['sofa']} | "
          f"Final Price: ${guest['final_price']:.2f}")
```
